## Introdução ao teste de hipótese

## Teste de hipótese com z para uma média amostral (unicaudal)

Por exemplo, suponhamos que uma professora redesenhe seu curso de história para que os alunos façam testes frequentes antes de fazer os exames. Depois de um semestre usando esses questionários, a nota média no exame final dos 25 alunos do seu curso foi $\bar{x} = 80$. A professora sabe que a nota média nesse mesmo exame final para todos os alunos que fizeram o curso com ela antes de ela exigir provas frequentes foi de 75 pontos com desvio padrão de 10 pontos ($\mu = 75, \sigma = 10$).
A professora tem todos os motivos para acreditar que os 25 alunos do novo curso de história são semelhantes a todos os alunos que ela ensinou no passado. Então, logicamente, todos os ex-alunos são considerados uma população e os 25 novos alunos são considerados uma amostra. Embora esses 25 novos alunos não tenham sido amostrados literalmente aleatoriamente da população anterior, é razoável esperar que obtenham pontuações semelhantes a todos os alunos anteriores que fizeram o mesmo exame. Com o teste de hipóteses, você pode determinar se a amostra de alunos forçados a responder aos testes frequentes teve pontuação "significativamente" melhor do que a dos alunos anteriores que não responderam aos testes. Se os questionários ajudassem a amostra, talvez exigir questionários para todos os futuros alunos também os ajudasse. Os questionários frequentes ajudaram? Se a nota média do exame final da amostra for melhor que a nota média do exame final da população, a diferença é criada pelo questionário ou por erro amostral?
Obviamente, a média amostral de 80 é numericamente maior do que a média populacional de 75. A questão da professora é se o desvio entre a média amostral de 80 e a média populacional de 75 é provável ou improvável de ter ocorrido devido a erro amostral. Você pode usar o processo de teste de hipótese e o z para uma média amostral que você aprendeu no capítulo anterior para determinar se a diferença de 5 pontos provavelmente se deve a um erro amostral ou se é provável que tenha sido criada pelo uma fonte de erro introduzida pelo uso do questionário. Os cálculos necessários para este teste de hipótese são idênticos aos que você fez no capítulo anterior. No entanto, precisamos introduzir alguns conceitos para avaliar hipóteses.

Esses novos conceitos são descritos abaixo em uma série de seis etapas.


_Etapa 1: examinar as variáveis para avaliar suposições estatísticas_

Todos os testes de hipóteses são baseados em suposições específicas e, se essas suposições forem violadas, esses testes podem produzir resultados enganosos. Portanto, o primeiro passo ao conduzir qualquer teste de hipótese é determinar se os dados que você está prestes a analisar atendem às suposições necessárias. Felizmente, se você compreender quatro suposições básicas, será capaz de executar com segurança todas as estatísticas de testes de hipóteses discutidas neste texto. Os quatro pressupostos básicos são (1) independência dos dados, (2) medição apropriada das variáveis para análise, (3) normalidade das distribuições e (4) homogeneidade da variância.

A independência da suposição de dados significa que a pontuação de cada participante dentro de uma condição é independente das pontuações de todos os outros participantes dentro da mesma condição. Por exemplo, nos testes de estudo, se testes frequentes melhorarem as notas dos alunos nos exames, você deverá garantir que todos os alunos responderam sozinhos a todas as perguntas do exame; se um ou mais alunos copiassem as respostas de outro aluno em vez de fornecerem as suas próprias respostas, esta fraude violaria o pressuposto da independência dos dados porque as respostas de um aluno "influenciavam" as respostas de outro aluno. Se essa trapaça fosse generalizada, isso arruinaria todo o estudo. Ao avaliar a independência dos dados, você está realmente avaliando o controle processual utilizado para coletar os dados. Se o controle processual de um estudo for muito fraco (por exemplo, houve trapaça generalizada ou algum outro problema com a coleta de dados), os dados violam a suposição de independência e você deverá refazer o estudo com mais cuidado.
Ao realizar um teste de hipótese usando um z para uma média amostral, a suposição apropriada de medição de variáveis significa que a variável independente (VI) deve identificar um grupo de pessoas que são diferentes da população de alguma forma, e a variável dependente (VD) deve ser medida em uma escala númerica quantitativa. Se o VI ou o VD não atenderem a esses requisitos, então o z para uma média amostral não é o teste apropriado. Por exemplo, no estudo de questionários frequentes, a VI é se os participantes responderam ou não aos questionários frequentes antes de fazerem o exame final. A VI identifica como a amostra e a população são diferentes – ou seja, a amostra respondeu a questionários frequentes e a população não. Ao realizar um z para uma média amostral, a VI é uma "variável de agrupamento"; identifica um grupo de pessoas que são diferentes da população. O VD neste estudo, as notas dos alunos nos exames finais, deve ser medido para que o desempenho de cada aluno possa ser quantificado com precisão. Em outras palavras, a VD deve ser medido como uma variável quantitativa. Se as notas do exame forem registradas como número de respostas corretas ou porcentagem de acertos, essa suposição seria atendida porque saberíamos exatamente o quanto melhor (ou pior) cada aluno se saiu em relação aos outros alunos. 
A suposição de normalidade das distribuições significa que a distribuição das médias amostrais para cada condição deve ter uma forma normal. Como você deve lembrar do capítulo anterior, o teorema central do limite nos diz que uma distribuição de médias amostrais terá uma forma normal se (a) a população original de variáveis tiver uma forma normal ou (b) o tamanho da amostra	usado no cálculo do estudo é 30 ou maior. Assim, estritamente falando, neste estudo sobre questionários frequentes, o professor deve determinar se a população original das variáveis de todos os alunos tem uma forma normal ou aumentar o tamanho da amostra de 25 para 30 para estar relativamente confiante de que a suposição de normalidade é válida. Contudo, na prática, a menos que os investigadores saibam que uma distribuição de variáveis é muito distorcida, normalmente procedem com tamanhos de amostra "próximos de" 30.
Em última análise, a solidez desta decisão baseia-se no quanto a forma de uma distribuição se desvia do normal. Se se desviar apenas ligeiramente da forma normal, um tamanho de amostra "próximo de" 30 é provavelmente suficiente. Contudo, se a forma de uma população se desviar substancialmente do normal, poderá ser necessária uma amostra ainda maior que 30. No final, os investigadores têm de "conhecer as suas variáveis" e fazer julgamentos sólidos. Para este estudo, o professor analisou a distribuição das notas de todos os ex-alunos e seu formato era normal. Portanto, a distribuição das médias amostrais terá uma forma normal, independentemente do tamanho da amostra. 
Ao realizar um z para uma média amostral, a suposição de homogeneidade de variância significa que as variâncias em cada condição do estudo são semelhantes. No estudo de questionários frequentes, isso significa que a variância das pontuações quando os alunos experimentam questionários frequentes precisa ser semelhante à variância das pontuações quando os alunos não experimentam questionários frequentes. Os investigadores não podem verificar esta suposição antes de recolherem os dados. Se o desvio padrão da amostra for o dobro do da população ou vice-versa, esta suposição pode ter sido violada. Neste estudo, o desvio padrão da amostra foi de 9.5 e o desvio padrão da população foi de 10. Assim, o pressuposto de homogeneidade de variância foi atendido.

_Etapa 2: expor as hipóteses nulas e de pesquisa Simbolicamente e Verbalmente_
         	
A segunda etapa no teste de hipóteses é estabelecer as hipóteses nula e alternativa (ou de pesquisa). Neste exemplo, o professor prevê que questionários frequentes aumentarão as pontuações dos testes. Assim, a sua hipótese de investigação afirma que os alunos que realizam testes frequentes terão uma média superior à população de alunos que não realizam testes frequentes. A representação simbólica formal da hipótese de pesquisa é sempre em termos de parâmetros populacionais. Neste caso, é:

$\mu_{quiz} > 75$ 

O "$\mu_{quiz} > 75$" representa a pontuação média do teste da população se todos eles respondessem a questionários frequentes antes do teste. Assim, a hipótese de investigação de "$\mu_{quiz} > 75$" prevê que se a população respondesse aos questionários, a sua pontuação média no teste seria maior do que é agora, 75. De forma menos formal, a hipótese de investigação prevê que o tratamento do questionário funcionará. Os questionários aumentarão as pontuações dos testes. 
Embora a hipótese de pesquisa preveja que os questionários aumentarão os resultados dos testes, é possível que os questionários não funcionem. Esta segunda possibilidade é chamada de hipótese nula e é sempre o oposto da hipótese de pesquisa. A representação simbólica formal da hipótese nula é

$\mu_{quiz} = 75$

Observe que a hipótese nula inclui o sinal de igual, enquanto a pesquisa não. A hipótese nula de "$\mu_{quiz} = 75$" prevê que se a população respondesse aos questionários, sua pontuação média no teste seria igual ao que é agora, 75. Menos formalmente, a hipótese nula prevê que o tratamento de questionários frequentes não funciona.
Você deve reconhecer que as hipóteses nula e de pesquisa são opostas e, portanto, mutuamente exclusivas. Eles abrangem coletivamente todos os resultados possíveis. Apenas uma dessas hipóteses pode estar correta. O objetivo de fazer testes de hipóteses é determinar qual dessas duas hipóteses tem maior probabilidade de ser verdadeira. As representações simbólicas e verbais de ambas as hipóteses são apresentadas na Tabela 6.1.

**Tabela 6.1 Representações Simbólicas e Verbais das Hipóteses Nulas e de Pesquisa para z de uma Média Amostral**

Tipo de Hipótese | Simbólico | Vebal | Diferença entre médias amostral e populacional
-- | -- | -- | --
Hipótese nula | $\mu_{quiz}=75;$ | Com questiónario notas $\approx$ 75 | Erro amostral
Hipótese de pesquisa | $\mu_{quiz}>75.$ | Com questiónario notas $>$ 75 | Intervenção melhora resultados

Contudo, é importante observar o subscrito "quiz" em $\mu_{quiz} > 75$. O subscrito refere-se àqueles que fazem testes frequentes antes do teste. Não sabemos qual seria a pontuação média do teste se toda a população de alunos respondesse aos testes antes do teste (ou seja, não sabemos o $\mu_{quiz}$). Sabemos que a média da população é 75 quando os alunos fazem a prova sem responder previamente aos questionários. A questão da pesquisa é se a realização de questionários frequentes criaria uma média populacional superior a 75. Portanto, a hipótese da pesquisa prevê que, se houvesse uma população de alunos que responderam aos testes antes do teste, a média do teste populacional ($\mu_{quiz}$) seria superior a 75 (ou seja, a média atual daqueles que não responderam aos testes antes do teste). Em contraste, a hipótese nula prevê que, se houvesse uma população de alunos que responderam aos testes antes do teste, a média do teste populacional ($\mu_{quiz}$) seria inferior a 75 (ou seja, a média atual daqueles que não responderam quiz antes de fazer o teste).

Neste cenário, a hipótese da pesquisa é essencialmente dizer que o aumento de 5 pontos na pontuação do exame, de 75 para 80, foi criado pelos questionários frequentes. Em contraste, a hipótese nula afirma que este aumento foi criado por erro amostral; questionários frequentes não levarão a um aumento de toda a população.

Hipóteses: 

$H_0: \mu_{quiz}=75;$

$H_a: \mu_{quiz}>75.$

_Passo 3: Definir as Regiões Críticas_

A terceira etapa no processo de teste de hipóteses é definir a região crítica. A região crítica é a coleção de todos os valores de z que são tão raros se o nulo for verdadeiro que, se qualquer um desses valores de z ocorrer, sugere que a hipótese nula é provavelmente falsa. Para entender a região crítica, precisamos retornar à distribuição das médias amostrais.

Figura 6.1 - Uma distribuição de médias amostrais com uma linha numérica na escala z (superior) e uma média amostral



A Figura 6.1 é a distribuição das médias amostrais para o estudo de questionário frequente. Contém todas as médias amostrais possíveis de tamanho N = 25 retiradas da população de alunos que realizaram o exame final sem completar questionários frequentes. O teorema do limite central nos diz que a média desta distribuição é 75 ($\mu$), e que o erro padrão da média é 2 ($SEM = \sigma/N = 10/25$). Assim, a distribuição é rotulada como com média 75 e desvio padrão 2. Em última análise, trabalharemos com valores de z, então também converteram cada um dos valores médios amostrais possíveis no eixo x em z (exemplo $z = (\bar{x} - \mu)/SEM = (77 - 75)/2 = 1$).
Se a hipótese nula for verdadeira (questionários frequentes não têm efeito no desempenho do teste), você esperaria um valor de z igual a 0 porque seria esperado que M e $\mu$ fossem idênticos (ou seja, seria esperado que ambos fossem 75). Portanto, o numerador da	fórmula z seria $\bar{x} - \mu = 0$. Mas, mesmo que a hipótese nula fosse verdadeira, o numerador da fórmula z pode não ser exatamente 0 porque é provável que haja algum erro de amostragem. Por exemplo, devido a um erro de amostragem, você não ficaria surpreso se obtivesse um valor z igual a 0.3, mesmo que o questionário não tenha efeito no desempenho. Você ficaria, no entanto, surpreso se obtivesse uma pontuação z de 2.9 se o questionário não tivesse efeito no desempenho.

```python
# HIDE CODE
from scipy.stats import norm
import numpy as np
from matplotlib import pyplot as plt

# Nomal com média = 75 e desvio padrão = 2
# Com barra horizontal na escala z
# simular valores dentro da distribuição normal N(0,1)
x=np.linspace(0-3*1, 0+3*1,1000)

Z = norm(75,2)

# simular valores dentro da distribuição normal N(0,1)
x=np.linspace(75-3*2, 75+3*2,1000)

ax1 = plt.subplot(1,1,1)

ax1.plot(x, Z.pdf(x), color='b')
ax1.set_xlabel('Nota')
ax1.set_ylabel('Densidade')
ax1.set_title('Distribuição de notas')
#plt.grid(True)

# Set scond x-axis
ax2 = ax1.twiny()

# Decide the ticklabel position in the new x-axis,
# then convert them to the position in the old x-axis
newlabel = [-2.5,-2,-1,0,1,2.5] # labels of the xticklabels: the position in the new x-axis
k2degc = lambda x: (x*2)+75 # convert function: from Kelvin to Degree Celsius
newpos   = [k2degc(t) for t in newlabel]   # position of the xticklabels in the old x-axis
ax2.set_xticks(newpos)
ax2.set_xticklabels(newlabel)

ax2.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
ax2.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
ax2.spines['bottom'].set_position(('outward', 36))
ax2.set_xlabel('Escala Z')
ax2.set_xlim(ax1.get_xlim())

# Save the figure
#plt.savefig('two_xticks_under.png', bbox_inches='tight', pad_inches=0.02, dpi=150)
```
