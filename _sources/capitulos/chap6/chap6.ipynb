## Introdução ao teste de hipótese

## Teste de hipótese com z para uma média amostral (unicaudal)

Por exemplo, suponhamos que uma professora redesenhe seu curso de história para que os alunos façam testes frequentes antes de fazer os exames. Depois de um semestre usando esses questionários, a nota média no exame final dos 25 alunos do seu curso foi $\bar{x} = 80$. A professora sabe que, antes de ela exigir testes frequentes, a nota média nesse mesmo exame final havia sido de 75 pontos, com desvio padrão de 10 pontos ($\mu = 75, \sigma = 10$).
A professora tem todos os motivos para acreditar que os 25 alunos do novo curso de história são semelhantes a todos os alunos que ela ensinou no passado. Então, logicamente, todos os ex-alunos são considerados uma população e os 25 novos alunos são considerados uma amostra. Embora esses 25 novos alunos não tenham sido amostrados literalmente de forma aleatória da população anterior, é razoável esperar que obtenham pontuações semelhantes a todos os alunos anteriores que fizeram o mesmo exame. Com o teste de hipóteses, você pode determinar se a amostra de alunos forçados a responder aos testes frequentes teve pontuação "significativamente" melhor do que a dos alunos anteriores que não responderam aos testes. Se os questionários ajudassem a amostra, talvez exigir questionários para todos os futuros alunos também os ajudasse. Os questionários frequentes ajudaram? Se a nota média do exame final da amostra for melhor que a nota média do exame final da população, a diferença terá sido criada pelo questionário ou por erro amostral?
Obviamente, a média amostral de 80 é numericamente maior do que a média populacional de 75. A questão da professora é se é provável ou improvável que o desvio entre a média amostral de 80 e a média populacional de 75 tenha ocorrido devido a erro amostral. Você pode usar o processo de teste de hipótese e o z para uma média amostral que você aprendeu no capítulo anterior para determinar se a diferença de 5 pontos provavelmente se deve a um erro amostral ou se, na verdade, essa diferença provavelmente foi criada pelo uso do questionário. Os cálculos necessários para este teste de hipótese são idênticos aos que você fez no capítulo anterior. No entanto, precisamos introduzir alguns conceitos para avaliar hipóteses.

Esses novos conceitos são descritos abaixo em uma série de seis etapas.


### _Etapa 1: examinar as variáveis para avaliar suposições estatísticas_

Todos os testes de hipóteses são baseados em suposições específicas e, se essas suposições forem violadas, esses testes podem produzir resultados enganosos. Portanto, o primeiro passo ao conduzir qualquer teste de hipótese é determinar se os dados que você está prestes a analisar atendem às suposições necessárias. Felizmente, se você compreender quatro suposições básicas, será capaz de executar com segurança todas as estatísticas de testes de hipóteses discutidas neste texto. Os quatro pressupostos básicos são (1) independência dos dados, (2) medição apropriada das variáveis para análise, (3) normalidade das distribuições e (4) homogeneidade da variância.

A suposição de independência dos dados significa que a pontuação de cada participante dentro de uma condição é independente das pontuações de todos os outros participantes dentro da mesma condição. Por exemplo, nos testes de estudo, se testes frequentes melhorarem as notas dos alunos nos exames, você deverá garantir que todos os alunos responderam sozinhos a todas as perguntas do exame; se um ou mais alunos copiassem as respostas de outro aluno em vez de fornecerem as suas próprias respostas, esta fraude violaria o pressuposto da independência dos dados porque as respostas de um aluno "influenciariam" as respostas de outro aluno. Se essa trapaça fosse generalizada, isso arruinaria todo o estudo. Ao avaliar a independência dos dados, você está, na verdade realmente avaliando o controle processual utilizado para coletar os dados. Se o controle processual de um estudo for muito fraco (por exemplo, houve trapaça generalizada ou algum outro problema com a coleta de dados), os dados violam a suposição de independência e você deverá refazer o estudo com mais cuidado.
Ao realizar um teste de hipótese usando um z para uma média amostral, a suposição apropriada de medição de variáveis significa que a variável independente (VI) deve identificar um grupo de pessoas que são diferentes da população de alguma forma, e a variável dependente (VD) deve ser medida em uma escala numérica quantitativa. Se a VI ou a VD não atenderem a esses requisitos, então o z para uma média amostral não é o teste apropriado. Por exemplo, no estudo de questionários frequentes, a VI representa os participantes que responderam ou não aos questionários frequentes antes de fazerem o exame final. A VI identifica como a amostra e a população são diferentes – ou seja, a amostra respondeu a questionários frequentes e a população não. Ao realizar um z para uma média amostral, a VI é uma "variável de agrupamento"; identifica um grupo de pessoas que são diferentes da população. A VD neste estudo, as notas dos alunos nos exames finais, deve ser medida para que o desempenho de cada aluno possa ser quantificado com precisão. Em outras palavras, a VD deve ser medida como uma variável quantitativa. Se as notas do exame forem registradas como número de respostas corretas ou porcentagem de acertos, essa suposição seria atendida porque saberíamos exatamente o quão melhor (ou pior) cada aluno se saiu em relação aos outros alunos. 
A suposição de normalidade das distribuições significa que a distribuição das médias amostrais para cada condição deve ter uma forma normal. Como você deve lembrar do capítulo anterior, o teorema central do limite nos diz que uma distribuição de médias amostrais terá uma forma normal se (a) a população original de variáveis tiver uma forma normal ou (b) o tamanho da amostra	usado no cálculo do estudo é 30 ou maior. Assim, estritamente falando, neste estudo sobre questionários frequentes, o professor deve determinar se a população original das variáveis de todos os alunos tem uma forma normal ou aumentar o tamanho da amostra de 25 para 30 para estar relativamente confiante de que a suposição de normalidade é válida. Contudo, na prática, a menos que os investigadores saibam que uma distribuição de variáveis é muito distorcida, normalmente procedem com tamanhos de amostra "próximos de" 30.
Em última análise, a solidez desta decisão baseia-se no quanto a forma de uma distribuição se desvia do normal. Se se desviar apenas ligeiramente da forma normal, um tamanho de amostra "próximo de" 30 é provavelmente suficiente. Contudo, se a forma de uma população se desviar substancialmente da normal, poderá ser necessária uma amostra ainda maior que 30. No final, os investigadores têm de "conhecer as suas variáveis" e fazer julgamentos sólidos. Para este estudo, o professor analisou a distribuição das notas de todos os ex-alunos e seu formato era normal. Portanto, a distribuição das médias amostrais terá uma forma normal, independentemente do tamanho da amostra. 
Ao realizar um z para uma média amostral, a suposição de homogeneidade de variância significa que as variâncias em cada condição do estudo são semelhantes. No estudo dos questionários frequentes, isso significa que a variância das pontuações, quando os alunos experimentam questionários frequentes, precisa ser semelhante à variância das pontuações de quando os alunos não os realizam. Os investigadores não podem verificar essa suposição antes de recolherem os dados. Se o desvio padrão da amostra for o dobro do da população ou vice-versa, essa suposição pode ter sido violada. Neste estudo, o desvio padrão da amostra foi de 9.5 e o desvio padrão da população foi de 10. Assim, o pressuposto de homogeneidade de variância foi atendido.

### _Etapa 2: expor as hipóteses nulas e de pesquisa Simbolicamente e Verbalmente_
         	
A segunda etapa no teste de hipóteses é estabelecer as hipóteses nula e alternativa (ou de pesquisa). Neste exemplo, o professor prevê que questionários frequentes aumentarão as pontuações dos exames. Assim, a sua hipótese de investigação afirma que os alunos que realizam testes frequentes terão uma média superior à população de alunos que não realizam testes frequentes. A representação simbólica formal da hipótese de pesquisa é sempre em termos de parâmetros populacionais. Neste caso, é:

$\mu_{quiz} > 75$ 

O "$\mu_{quiz} > 75$" representa a pontuação média do exame final da população se todos eles respondessem a questionários frequentes antes do exame. Assim, a hipótese de investigação de "$\mu_{quiz} > 75$" prevê que se a população respondesse aos questionários, a sua pontuação média no exame seria maior do que é agora, 75. De maneira menos formal, a hipótese de investigação prevê que a estratégia de aplicar os questionários funcionará. Os questionários aumentarão as pontuações dos exames. 
Embora a hipótese de pesquisa preveja que os questionários aumentarão os resultados dos testes, é possível que os questionários não funcionem. Essa segunda possibilidade é chamada de hipótese nula e é sempre o oposto da hipótese de pesquisa. A representação simbólica formal da hipótese nula é

$\mu_{quiz} = 75$

Observe que a hipótese nula inclui o sinal de igual, enquanto a da pesquisa, não. A hipótese nula de "$\mu_{quiz} = 75$" prevê que, se a população respondesse aos questionários, sua pontuação média no exame seria igual ao que é agora, 75. Menos formalmente, a hipótese nula prevê que o tratamento de questionários frequentes não funciona.
Você deve reconhecer que as hipóteses nula e de pesquisa são opostas e, portanto, mutuamente exclusivas. Elas abrangem coletivamente todos os resultados possíveis. Apenas uma dessas hipóteses pode estar correta. O objetivo de fazer testes de hipóteses é determinar qual dessas duas hipóteses tem maior probabilidade de ser verdadeira. As representações simbólicas e verbais de ambas as hipóteses são apresentadas na Tabela 6.1.

**Tabela 6.1 Representações Simbólicas e Verbais das Hipóteses Nulas e de Pesquisa para z de uma Média Amostral**

Tipo de Hipótese | Simbólico | Vebal | Diferença entre médias amostral e populacional
-- | -- | -- | --
Hipótese nula | $\mu_{quiz}=75;$ | Com questiónario notas $\approx$ 75 | Erro amostral
Hipótese de pesquisa | $\mu_{quiz}>75.$ | Com questiónario notas $>$ 75 | Intervenção melhora resultados

Alguns alunos acham a notação simbólica das hipóteses nula e de pesquisa um pouco confusa porque estamos tentando determinar se $\mu_{quiz} > 75$, e já sabemos que $\bar{x} = 80$ e $\mu = 75$, com base nas informações fornecidas no problema. Contudo, é importante observar o subscrito "quiz" em $\mu_{quiz} > 75$. O subscrito refere-se àqueles que fazem testes frequentes antes do exame. Não sabemos qual seria a pontuação média do exame se toda a população de alunos respondesse aos testes antes do exame final (ou seja, não sabemos o $\mu_{quiz}$). Sabemos que a média da população é 75 quando os alunos fazem a prova sem responder previamente aos questionários. A questão da pesquisa é se a realização de questionários frequentes criaria uma média populacional superior a 75. Portanto, a hipótese da pesquisa prevê que, se houvesse uma população de alunos que responderam aos testes antes da prova, a média populacional do exame ($\mu_{quiz}$) seria superior a 75 (ou seja, maior que a média atual daqueles que não responderam aos testes antes da prova). Em contraste, a hipótese nula prevê que, se houvesse uma população de alunos que responderam aos testes antes da prova final, a média populacional do exame ($\mu_{quiz}$) seria inferior a 75 (ou seja, a média atual daqueles que não responderam testes antes de fazer o exame).

Neste cenário, a hipótese da pesquisa é essencialmente dizer que o aumento de 5 pontos na pontuação do exame, de 75 para 80, foi criado pelos questionários frequentes. Em contraste, a hipótese nula afirma que esse aumento foi criado por erro amostral; questionários frequentes não levarão a um aumento de toda a população.

Hipóteses: 

$H_0: \mu_{quiz}=75;$

$H_a: \mu_{quiz}>75.$

### _Passo 3: Definir as Regiões Críticas_

A terceira etapa no processo de teste de hipóteses é definir a região crítica. A região crítica é a coleção de todos os valores de z que são tão raros no caso de a hipótese nula ser verdadeira que, se qualquer um desses valores de z ocorrer, isso sugere que a hipótese nula é provavelmente falsa. Para entender a região crítica, precisamos retornar à distribuição das médias amostrais.

Figura 6.1 - Uma distribuição de médias amostrais com uma linha numérica na escala z (superior) e uma média amostral

A Figura 6.1 é a distribuição das médias amostrais para a aplicação dos questionários frequentes. Ela contém todas as médias amostrais possíveis de tamanho N = 25 retiradas da população de alunos que realizaram o exame final sem completar questionários frequentes. O teorema central do limite nos diz que a média desta distribuição é 75 ($\mu$), e que o erro padrão da média é 2 ($SEM = \sigma/\sqrt{N} = 10/\sqrt{25}$). Assim, a distribuição é rotulada como com média 75 e desvio padrão 2. Em última análise, trabalharemos com valores de z, então também convertemos cada um dos valores médios amostrais possíveis no eixo x em z (exemplo $z = (\bar{x} - \mu)/SEM = (77 - 75)/2 = 1$).
Se a hipótese nula for verdadeira (questionários frequentes não têm efeito no desempenho do exame), você esperaria um valor de z igual a 0 porque seria esperado que $\bar{x}$ e $\mu$ fossem idênticos (ou seja, seria esperado que ambos fossem 75). Portanto, o numerador da	fórmula z seria $\bar{x} - \mu = 0$. Mas, mesmo que a hipótese nula fosse verdadeira, o numerador da fórmula z pode não ser exatamente 0 porque é provável que haja algum erro de amostragem. Por exemplo, devido a um erro de amostragem, você não ficaria surpreso se obtivesse um valor z igual a 0.3, mesmo que o questionário não tenha efeito no desempenho. Você ficaria, no entanto, surpreso se obtivesse uma pontuação z de 2.9 se o questionário não tivesse efeito no desempenho.

````python
# HIDE CODE
from scipy.stats import norm
import numpy as np
from matplotlib import pyplot as plt

# Nomal com média = 75 e desvio padrão = 2
# Com barra horizontal na escala z
# simular valores dentro da distribuição normal N(0,1)
x=np.linspace(0-3*1, 0+3*1,1000)

Z = norm(75,2)

# simular valores dentro da distribuição normal N(0,1)
x=np.linspace(75-3*2, 75+3*2,1000)

ax1 = plt.subplot(1,1,1)

ax1.plot(x, Z.pdf(x), color='b')
ax1.set_xlabel('Nota')
ax1.set_ylabel('Densidade')
ax1.set_title('Distribuição de notas')
#plt.grid(True)

# Set scond x-axis
ax2 = ax1.twiny()

# Decide the ticklabel position in the new x-axis,
# then convert them to the position in the old x-axis
newlabel = [-2.5,-2,-1,0,1,2.5] # labels of the xticklabels: the position in the new x-axis
k2degc = lambda x: (x*2)+75 # convert function: from Kelvin to Degree Celsius
newpos   = [k2degc(t) for t in newlabel]   # position of the xticklabels in the old x-axis
ax2.set_xticks(newpos)
ax2.set_xticklabels(newlabel)

ax2.xaxis.set_ticks_position('bottom') # set the position of the second x-axis to bottom
ax2.xaxis.set_label_position('bottom') # set the position of the second x-axis to bottom
ax2.spines['bottom'].set_position(('outward', 36))
ax2.set_xlabel('Escala Z')
ax2.set_xlim(ax1.get_xlim())

# Save the figure
#plt.savefig('two_xticks_under.png', bbox_inches='tight', pad_inches=0.02, dpi=150)
````

Embora a Figura 6.1 dê uma ideia geral de quão prováveis serão os valores z específicos se a hipótese nula for verdadeira, a tabela de probabilidades normais fornece probabilidades muito mais precisas. Por exemplo, a probabilidade de obter uma média amostral com o valor de z igual a 1 ou superior é 0.1587 (ou seja, a área da cauda de z = 1) se a hipótese nula for verdadeira. A probabilidade de obter uma média amostral com valor de z igual a 2.5 ou superior é 0.0062 se a hipótese nula for verdadeira. Claramente,se a hipótese nula for verdadeira, um valor z de 2.5 é menos provável do que um valor z de 1. No entanto, o teste de hipóteses requer um ponto de corte que identifique claramente quais pontuações z são improváveis se a hipótese nula for verdadeira. O ponto de corte mais comumente usado é uma probabilidade de 0.05. Por isso, valores z com probabilidades de 0.05 ou inferiores são consideradas improváveis de ocorrer se a hipótese nula for verdadeira. É improvável que valores de z com probabilidades inferiores a 0.05 sejam consequência de erros de amostragem, portanto é mais provável que decorram de outra coisa (por exemplo, questionários frequentes que melhoram as pontuações dos exames). Por outro lado, valores de z com probabilidades superiores a 0.05 são considerados relativamente prováveis de ocorrer se a hipótese nula for verdadeira; o desvio desses valores de z de 0 provavelmente se deve a um erro de amostragem. 	
É importante lembrar que você pode obter valores de z com uma probabilidade menor que o limite de probabilidade de 0.05 simplesmente devido a um erro de amostragem. Na verdade, se a hipótese nula for verdadeira, você teria 5% de chance de obter um valor z extremo, o que o levaria a concluir incorretamente que os questionários frequentes melhoraram os resultados dos testes. Este tipo de erro é denominado erro do Tipo I; explicaremos esse erro com mais detalhes posteriormente neste capítulo. Quando você decide usar 0,05 como ponto de corte e a hipótese nula é realmente verdadeira, há 5% de chance de obter uma pontuação z extrema simplesmente devido a um erro de amostragem.

O valor de corte de 0.05 usado para identificar quais valores z são improváveis se a hipótese nula for verdadeira é chamado de nível alfa ($\alpha$). Esse ponto de corte de nível alfa também pode ser convertido em um ponto de corte de valor z procurando 0.05 na coluna final da tabela normal e encontrando o valor de z com uma cauda de 0.05. A Figura 6.2 ilustra como isso é feito. 


````python
# HIDE CODE
import pandas as pd
import numpy as np
from scipy import stats

Z = stats.norm(0,1)

zs = np.arange(1.63, 1.69, 0.01)
bodies = [round(Z.cdf(x), 4) for x in zs]
tails = [1-x for x in bodies]

df = pd.DataFrame(zip(zs, bodies, tails), columns=['z', 'Corpo', 'Cauda'])

color = (df.Cauda == df.loc[2,'Cauda']).map({True: 'background-color: yellow', False: ''})

df.style.apply(lambda s: color)
````

Como você pode ver, a tabela não inclui uma probabilidade final de exatamente 0.05, mas há duas probabilidades próximas de 0.05: 0.0505 e 0.0495. Na Figura 6.2, escolhemos 0.0495 e a utilizamos para encontrar o ponto de corte do valor z de 1.65. Forçado a escolher entre 0.0505 e 0.0495, a maioria dos estatísticos recomendaria o uso de 0.0495 porque garante que a taxa de erro Tipo I não seja superior a 5%, em vez de ligeiramente superior a 5%. A mensagem para levar para casa é que o ponto de corte da pontuação z que usaremos para um teste de significância unilateral de 0.05 será 1.65 neste curso.
A Figura 6.3 pode ajudar a explicar a relação entre o valor alfa e o valor crítico. O valor alfa de 0.05 é representado pela área sombreada na extremidade direita da curva na Figura 6.3. A área sombreada representa os 5% superiores da curva. O valor localizado no início da área improvável do valor de z (a área sombreada) é chamado de valor crítico. Neste caso, o valor crítico de z é 1.65. O valor crítico é o ponto de corte do valor de z; valores de z iguais ou maiores que o valor crítico são considerados improváveis de ocorrer se a hipótese nula for verdadeira. Em contraste, valores z mais próximos de zero do que o valor crítico (1.65) são considerados prováveis de ocorrer se a hipótese nula for verdadeira.

````python
# HIDE CODE
# simular valores dentro da distribuição normal N(0,1)
import matplotlib.pyplot as plt

x=np.linspace(0-3*1, 0+3*1,1000)

plt.plot(x, Z.pdf(x), color='b')

zc = 1.65

px=np.arange(zc,0+3*1,0.1)
plt.fill_between(px, Z.pdf(px),color='r')
plt.title('Distribuição Z com área acima de 1.65 destacada.')
#plt.savefig('regiao_critica.png', bbox_inches='tight', pad_inches=0.02, dpi=150)
````

É importante notar que a hipótese de investigação para o exemplo dos questionários frequentes prevê que as pontuações serão mais elevadas quando se utilizam questionários frequentes do que quando não se utilizam questionários frequentes. Como a hipótese de pesquisa previu que a média amostral seria maior que a média populacional, o valor crítico é positivo (ou seja, +1.65). No entanto, se a hipótese de pesquisa tivesse previsto uma média amostral inferior à média populacional, o valor crítico seria negativo (ou seja, -1.65), e a área sombreada, chamada de região crítica, estaria na cauda negativa do valor acima da distribuição.

### _Etapa 4: calcular a estatística de teste (z para uma média amostral)_

_4a. Calcule o desvio observado entre a média amostral e a média da população_

Encontre o desvio entre a média amostral e a média populacional.

$\bar{x} - \mu = 80 - 75 = 5$

_4b. Calcule o desvio esperado devido ao erro de amostragem_

Calcule a quantidade esperada de erro amostral dado o tamanho da amostra.
                                        
$SEM_p = \frac{\sigma}{\sqrt{N}} = \frac{10}{\sqrt{25}} = 2$

### _Etapa 5: calcule o tamanho do efeito e descreva o seu grau (pequeno, médio ou grande)_

O objetivo do teste de hipótese é decidir se a diferença entre uma média amostral e uma média populacional pode ser atribuída a um efeito de tratamento ou se é devida a um erro amostral. Se a hipótese nula for rejeitada, o pesquisador está dizendo que a diferença observada provavelmente se deve a um efeito de tratamento e não a um erro de amostragem. Um tópico relacionado, mas diferente, é o tamanho da diferença entre a média da amostra e a média da população. Por exemplo, o pesquisador neste exemplo concluiu que os questionários frequentes parecem melhorar os resultados dos exames, mas até que ponto os questionários frequentes melhoraram os resultados dos exames finais? Essa questão não pode ser respondida com testes de hipóteses. Em vez disso, ela requer outro procedimento estatístico denominado tamanho do efeito. Um tamanho de efeito é um índice da diferença entre a média da amostra e a média da população. Neste caso, o tamanho do efeito indica quão grande é a diferença entre a média da amostra que respondeu aos questionários frequentes e a média da população que não respondeu aos questionários frequentes. A estatística de tamanho do efeito normalmente usada ao comparar duas médias é denotada por _d_. O numerador de d é igual ao z de uma média amostral (ou seja, a diferença entre as médias). O denominador, porém, é diferente. Quando se calcula o z para uma média amostral, você divide pelo erro padrão da média (ou seja, erro amostral típico). $d = \frac{Desvio\quad observado\quad entre\quad as\quad médias}{Desvio\quad padrão} = \frac{\bar{x}-\mu}{\sigma} = \frac{80-75}{10} = 0.5$

Ao calcular d, você pode descrever o tamanho relativo do desvio entre a média amostral e a média populacional. Um d = 0.5 significa que a média da amostra foi meio desvio padrão superior à média da população. Um d = 0.5 é bom? _Regras de Teste de Hipótese_

O processo de teste de hipótese nada mais é do que um conjunto formal de regras que os pesquisadores usam para determinar se a hipótese nula ($H_0$) é provável ou improvável. Se a hipótese nula for verdadeira, os pesquisadores esperam que a estatística computada esteja próxima de um valor específico. Ao realizar um z para uma média amostral, os pesquisadores esperam que o valor de z esteja próximo de 0 se a hipótese nula for verdadeira. Cada valor de z tem alguma probabilidade de ocorrer. Se a hipótese nula for verdadeira, então valores de z próximos de 0 têm uma probabilidade alta e valores de z distantes de 0 têm uma probabilidade baixa. Se a probabilidade de um valor de z for baixa o suficiente, os pesquisadores rejeitam a hipótese nula. A lógica desta decisão é que um valor de z terá uma probabilidade baixa se a hipótese nula for verdadeira e esse valor de z ainda acontecer, então a hipótese nula provavelmente não é verdadeira. Mas quão baixa deve ser a probabilidade de um valor z antes que a hipótese nula seja rejeitada? Deve ser inferior ao valor $\alpha$. No exemplo anterior, usamos um valor alfa de 0.05. Os pesquisadores normalmente escolhem entre um valor $\alpha$ de 0.05 (5%) ou 0.01  (1%). Em geral, se quiser minimizar os erros do Tipo I, escolha um alfa de 0.01 e, se quiser maximizar o poder estatístico, escolha um alfa de 0.05. Em ambos os casos, sempre que o valor p for menor que o nível alfa, é improvável que a hipótese nula seja verdadeira e você deve rejeitá-la.

O que é um valor p?

Um valor p é a probabilidade de observar o valor calculado para a estatística do teste ou um valor mais extremo assumindo que a hipótese nula é verdadeira. É usado por pesquisadores para determinar se devem ou não rejeitar uma hipótese nula.

A Figura 6.4 pode ajudá-lo a visualizar como os valores z obtidos, os valores críticos e os valores p estão relacionados.

Realizaremos vários procedimentos estatísticos diferentes neste livro (por exemplo, z, t, r, F e $\chi^2$). Para cada um desses testes estatísticos, as regras básicas do teste de hipóteses são as mesmas. Se o valor estatístico calculado (denominado valor obtido) for mais extremo que o valor crítico, deve-se rejeitar a hipótese nula. Além disso, se o valor p para um valor obtido for menor ou igual ao valor alfa (geralmente 0.05 ou 0.01), você deve rejeitar a hipótese nula.

Sempre que você for confrontado com a questão de determinar se deve ou não rejeitar a hipótese nula, poderá usar uma das duas regras a seguir:

1. Se o valor obtido for mais extremo que o valor crítico, você deve rejeitar a hipótese nula.
2. Se o valor p for menor que o valor alfa, você deve rejeitar a hipótese nula.

A Regra 1 sempre levará à mesma decisão que a Regra 2 e vice-versa. Essas duas regras são completamente dependentes. Elas são análogas a dois caminhos diferentes para o mesmo local. Não importa qual desses dois caminhos você tome – você sempre terminará no mesmo lugar (ou seja, na mesma decisão estatística).

_Por que os estatísticos "falharam em rejeitar a hipótese nula" em vez de "aceitar a hipótese nula"_

Os alunos muitas vezes se perguntam por que os estatísticos usam a dupla negativa confusa 'não rejeitar a hipótese nula' em vez da expressão mais fácil de entender 'aceitar a hipótese nula'. Ofereceremos duas explicações relacionadas. A primeira será no contexto de um  exemplo de pesquisa específico. A segunda será um argumento lógico mais abstrato.
Um pesquisador que testa a eficácia de um programa destinado a reduzir a ansiedade matemática administra o programa a uma amostra de N = 25 pessoas. Após os participantes concluírem o programa, o pesquisador compara sua pontuação média de ansiedade ($\bar{x} = 55$) com o parâmetro populacional conhecido ($\mu = 58, \sigma = 10$). O valor de z resultante para a média amostral de -1.5 não está mais distante de 0 do que o valor crítico de -1.65, então ele não consegue rejeitar a hipótese nula. A hipótese nula indica que o programa não é eficaz. Ao não rejeitar esta hipótese, o pesquisador está dizendo que o presente estudo não forneceu evidências suficientes para rejeitá-la. No entanto, ele também reconhece a possibilidade de que a hipótese nula possa não ter sido rejeitada devido a um ou mais possíveis problemas com o estudo.

Por exemplo, uma hipótese nula falsa pode não ser rejeitada por um grande número de razões, incluindo erro de amostragem, se o programa precisasse de mais tempo para ser eficaz ou se houvesse uma série de problemas metodológicos com o estudo. Ao 'não rejeitar a hipótese nula', o pesquisador está reconhecendo todas essas possibilidades. Assim, a frase 'não rejeitar a hipótese nula' significa que atualmente não existem provas suficientes para rejeitar a hipótese nula, mas reconhece que pesquisas futuras poderão fornecer essa evidência. Em contraste, a frase 'aceitar a hipótese nula' ignora a perspectiva de que pesquisas futuras possam mudar a nossa conclusão. 'Aceitar a hipótese nula' implica uma conclusão final e definitiva que desmente a maioria das situações de investigação. Ignorar possibilidades alternativas não é consistente com a natureza cautelosa das conclusões científicas.

A segunda explicação para o motivo pelo qual os investigadores não utilizam as palavras 'aceitar a hipótese nula' está relacionada com uma limitação inerente ao raciocínio indutivo. Por exemplo, suponha que você tenha a hipótese de que todos os cisnes são brancos. Você poderia testar sua hipótese indo a um lago local. Se você descobrir que todos os cisnes naquele lago são de fato brancos, sua visita ao lago local significa que você deveria 'aceitar' a hipótese de que todos os cisnes são brancos? Não, é possível que exista um lago diferente que tenha pelo menos um cisne negro. Por que os cientistas dizem "esta pesquisa sugere" em ver de "esta pesquisa prova"

Embora estejamos no tópico de como descrever as conclusões da pesquisa, vale a pena explicar por que os cientistas usam uma linguagem cautelosa, até mesmo experimental, ao descrever os resultados da pesquisa. Se você ouviu cientistas falarem no rádio ou na televisão ou se leu citações precisas de cientistas, provavelmente já os ouviu usar a frase "esta pesquisa sugere" em vez da frase mais impressionante "esta pesquisa prova". Por que os cientistas evitam intencionalmente a palavra prova? A razão está intimamente relacionada com a discussão na seção anterior. Por exemplo, suponhamos que um investigador conduziu um estudo no qual rejeitou uma hipótese nula e concluiu que um programa de tutoria aumenta as pontuações dos alunos em ciências num exame padronizado. Com base apenas nesse resultado, se o cientista dissesse: "Este estudo prova que este programa de tutoria funciona", estaria ignorando a possibilidade de o resultado do estudo ter sido devido a um erro amostral (sempre uma possibilidade). Ele também estaria ignorando a possibilidade de o resultado ter sido causado por alguma falha desconhecida na metodologia do estudo. Ao utilizar a linguagem mais experimental, "esta investigação sugere", o cientista reconhece que, embora a informação actual que possui sugira que o programa funciona, é possível que trabalhos futuros possam alterar essa conclusão. Uma das razões pelas quais o processo científico é tão poderoso é que está disposto a mudar a sua posição se novas evidências sugerirem que uma nova posição é mais precisa. Se alguém afirma ter a resposta final, você pode ter certeza de que essa pessoa não é um cientista.

Os cientistas sempre reconhecem que uma resposta pode mudar à medida que mais resultados de investigação se acumulam. Então, isso significa que nunca poderemos ter confiança em qualquer conclusão científica? Bem, não. Se um determinado resultado tiver sido replicado por muitos estudos utilizando diferentes metodologias de investigação, é muito improvável que essa conclusão seja revertida por pesquisas futuras. Por exemplo, muitos estudos de investigação diferentes realizados por muitas equipes de investigação diferentes, utilizando muitas metodologias diferentes, relataram que questionários frequentes (ou seja, testes) levam a melhores pontuações de memória (Rowland, 2014). Esta é uma descoberta tão comum que é extremamente improvável que seja revertida por pesquisas futuras. Embora um único estudo não possa provar uma conclusão, uma infinidade de estudos que sugerem a mesma conclusão deveria tornar-nos extremamente confiantes nessa conclusão.
